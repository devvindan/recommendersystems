```python
import numpy as np
import pandas as pd
import seaborn
```

```python
data = pd.read_excel("data.xls")
data
```

```python
# user correlations
correlations = pd.DataFrame(data.transpose(), data.columns, data.index).corr()
correlations
```

```python

seaborn.heatmap(correlations)
```

```python
neighbours_3867 = correlations[3867].sort_values(ascending=False)[1:6]
```

```python
neighbours_89 = correlations[89].sort_values(ascending=False)[1:6]
```

```python
neighbours_3712 = correlations[3712].sort_values(ascending=False)[1:6]
```

```python
neighbours_3712.index
```

```python
recommendations = data.fillna(0)
```

```python
recommendations.loc[neighbours_3712.index]
```

```python
(recommendations.loc[neighbours_3867.index].multiply(
    neighbours_3867, axis=0).sum(axis=0) / (recommendations.iloc[:, :] != 0).multiply(
    neighbours_3867, axis=0).sum(axis=0)).sort_values(ascending=False)
```

```python
(recommendations.loc[neighbours_89.index].multiply(
    neighbours_89, axis=0).sum(axis=0) / (recommendations.iloc[:, :] != 0).multiply(
    neighbours_89, axis=0).sum(axis=0)).sort_values(ascending=False)
```

```python
(recommendations.loc[neighbours_3712.index].multiply(
    neighbours_3712, axis=0).sum(axis=0) / (recommendations.iloc[:, :] != 0).multiply(
    neighbours_3712, axis=0).sum(axis=0)).sort_values(ascending=False)
```

```python
#3867 89
user_means = data.mean(axis=1)

#user_means.loc[3867]
((recommendations.subtract(user_means, axis=0).multiply((recommendations.iloc[:, :] != 0))).loc[neighbours_3712.index].multiply(
    neighbours_3712, axis=0).sum(axis=0) / (recommendations.iloc[:, :] != 0).multiply(
    neighbours_3712, axis=0).sum(axis=0) + user_means.loc[3712]).sort_values(ascending=False)
```

```python
((recommendations.subtract(user_means, axis=0).multiply((recommendations.iloc[:, :] != 0))).loc[neighbours_3867.index].multiply(
    neighbours_3867, axis=0).sum(axis=0) / (recommendations.iloc[:, :] != 0).multiply(
    neighbours_3867, axis=0).sum(axis=0) + user_means.loc[3867]).sort_values(ascending=False)
```

```python
((recommendations.subtract(user_means, axis=0).multiply((recommendations.iloc[:, :] != 0))).loc[neighbours_89.index].multiply(
    neighbours_89, axis=0).sum(axis=0) / (recommendations.iloc[:, :] != 0).multiply(
    neighbours_89, axis=0).sum(axis=0) + user_means.loc[89]).sort_values(ascending=False)
```
